```python
import pandas as pd
import requests
from config import api_key
import json
```

```python
# to find list of republicans, this search will be the same, 
# except the party param will equal "Rep" instead of "Dem"

url="https://api.open.fec.gov/v1/candidates/"
params={
    "has_raised_funds":True,
    "sort_null_only":False,
    "election_year":2018,
    
    "office":"S",
    "api_key":api_key,
    "party":"Dem"
}
data=requests.get(url=url,params=params).json()
```

```python
# the fec api limits to 20 results per page, so in order to get
# all results, we need to set up a data frame for each page of results.
# here is the df for page 1. in order to see how many pages of
# results there are total, i first ran this search in postman,
# which gave me the total number of pages for this list.

candidate_page_1=pd.DataFrame(data["results"])
candidate_page_1.head()
```

```python
# Here is the call for the second page. this is an exacy copy
# of the search above, except i added a "page" param, which
# is set to page 2. this will get results 21 - 40. 

url="https://api.open.fec.gov/v1/candidates/"
params_2={
    "has_raised_funds":True,
    "sort_null_only":False,
    "election_year":2018,
    "office":"S",
    "api_key":api_key,
    "party":"Dem",
    "page":2
}
page_2=requests.get(url=url,params=params_2).json()
```

```python
candidate_page_2=pd.DataFrame(page_2["results"])
candidate_page_2.head()
```

```python
# page 3 search. 

url="https://api.open.fec.gov/v1/candidates/"
params_3={
    "has_raised_funds":True,
    "sort_null_only":False,
    "election_year":2018,
    "office":"S",
    "api_key":api_key,
    "party":"Dem",
    "page":3
}
page_3=requests.get(url=url,params=params_3).json()
```

```python
# page 3 df
candidate_page_3=pd.DataFrame(page_3["results"])
candidate_page_3.head()
```

```python
# page 4

url="https://api.open.fec.gov/v1/candidates/"
params_4={
    "has_raised_funds":True,
    "sort_null_only":False,
    "election_year":2018,
    "office":"S",
    "api_key":api_key,
    "party":"Dem",
    "page":4
}
page_4=requests.get(url=url,params=params_4).json()
```

```python
candidate_page_4=pd.DataFrame(page_4["results"])
candidate_page_4.head()
```

```python
url="https://api.open.fec.gov/v1/candidates/"
params_5={
    "has_raised_funds":True,
    "sort_null_only":False,
    "election_year":2018,
    "office":"S",
    "api_key":api_key,
    "party":"Dem",
    "page":5
}
page_5=requests.get(url=url,params=params_5).json()
```

```python
candidate_page_5=pd.DataFrame(page_5["results"])
candidate_page_5.head()
```

```python
candidate_page_1.head()
```

```python
# all page dfs are added to a list so they can be combined
# into a single df via concatenation 

dfs_to_concat = [candidate_page_1, candidate_page_2, candidate_page_3, candidate_page_4, candidate_page_5]
```

```python
# new df created by concat'ing the list above
dem_candidates=pd.concat(dfs_to_concat)
```

```python
# final df for dems created by narrowing the concatenated df
# by the columns we want to keep

dem_candidates=dem_candidates[["name", "candidate_id", "state", "party", "incumbent_challenge_full", "office_full", "cycles", "election_years", "party", "candidate_status"]]
```

```python
# test the final df shape to make sure all records were
# successfully added. when i ran the postman call, i saw
# that the total list contained 81 records, so this df
# successfully contains all candidates
dem_candidates.reset_index()
```

```python
dem_candidates.to_csv("dem_candidate.csv")
```
